```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import random
from matplotlib.ticker import FuncFormatter
```

```python
#see proposal for definition of functions
def func_comission(deposit_pool, min_commission, max_commission, min_deposit, max_deposit):
    return max(min_commission,min(max_commission, min_commission + (max_commission-min_commission)/(max_deposit-min_deposit)*deposit_pool))

def func_share_no(deposit_pool, min_no_commission_share, max_no_commission_share, kink, max_deposit):
    return max(min_no_commission_share, min(max_no_commission_share, min_no_commission_share + (max_no_commission_share-min_no_commission_share)/(max_deposit*(1-kink))*(deposit_pool-max_deposit*kink)))

def func_collateral_need(num_validators,collateral_power,collateral_base):
    return (num_validators*collateral_base)**collateral_power

## for chart formatting

sns.set_theme()

# Function to format y-axis labels as percentages
def to_percent(decimal_places):
    def formatter(y, position):
        # Note: `position` is a required argument but not used here
        return f'{y * 100:.{decimal_places}f}%'
    return formatter


```

```python
#all data pulled via rocketscan.io
#load validatorcount per node operator
with open('rp-validators-per-node.txt', 'r') as f:
    num = f.readline().split(',')
    num_list = [x.replace('\'', '') for x in num]

lst_nodes_tmp = [int(x.replace('\'', '')) for x in num]
lst_nodes=[i for i in lst_nodes_tmp if i>0]
```

```python
### set design- and simulation parameters


## proposal related parameters
#total commission based on deposit pool size
min_commission = 0.05 
max_commission = 0.14
min_deposit = 0 
max_deposit = 18000

#NO share of commission based on deposit pool size
kink_no_share = 0.8 
min_no_commission_share = 0.2
max_no_commission_share = 0.7


#sublinear collateral requirement
collateral_base = 8
collateral_power= 2/3

#RPL stake requirement per validator
min_stake_requirement =  150



## other market/protocol inputs:
rpl_price_init = 25
eth_price_init = 3500

deposit_queue_init = 18000
rpl_supply_init = 20445460
rpl_stake_init = 10614717

num_validators_initial = sum(lst_nodes)
reth_tvl_initial = num_validators_initial*32
pool_size_eth = 4 #for comparisons with status quo/other proposals

#modeling assumptions
eth_yield = 0.04
additional_rpl_stake_init = 1

simulation_length = 365*3
validator_range=range(1,2000) #current max number of validators per single node operator is 1200
additional_eth_yield_performance_factor = 0.0007/0.061 #there was a 0.07% yield impovement when exlcuding low performers at eth_yield of 6.1%, so use that to scale expected improvements (based on https://dao.rocketpool.net/t/rapid-research-incubator-submission-reth-protection-through-rpl-rerouting-deflation/2599)

#we need to assume that staking stops at some stake ratio as 100% stake ratio is unrealistic
max_stake_ratio = 0.8 
stake_rate_deduction = 0.33

#for the projection based on deposit growth and price changes
growth_rate_tvl = 0.33
rpl_inflation=0.015
do_price_change = False #put true if you want RPL price vs. ETH to change
mu_rpl = 0.1
sigma_rpl = 0.05


#Inputs from original proposal
proposed_commission = 0.14
proposed_no_share = 0.035
proposed_stake_share = 0.05






```

## Commission function and resulting yields for RPL staking

### Commmssion function

```python
#assume stable values for 1 year, loop through different deposit pool sizes 

dct_out = {
    'deposit_pool':[],
    'commission':[],
    'share_no_commission':[],
    'eth_yield_no_commssion':[],
    'eth_yield_stakers_commission':[],
    'eth_yield_no_proposal':[],
    'rpl_yield_stakers':[],
    'rpl_yield_stakers_proposal':[],
    'share_stake_commission':[],
    'stake_commission':[],
    'no_commission':[],
    'reth_yield':[],
    'reth_yield_current':[],
    'reth_yield_improvement_commission':[],
    'reth_yield_improvement_performance':[]
}

for deposit_pool in np.arange(0,max_deposit*1.1,500):
        
        #total commission and NO- and RPL staker share of commission based on deposit pool size
        commission = func_comission(deposit_pool, min_commission, max_commission, min_deposit, max_deposit)
        share_no_commission = func_share_no(deposit_pool, min_no_commission_share, max_no_commission_share, kink_no_share, max_deposit)
        eth_yield_no_commssion = eth_yield*share_no_commission*commission
        eth_yield_stakers_commission=eth_yield*(1-share_no_commission)*commission
        
        #values for original proposal
        eth_yield_no_proposal=eth_yield*proposed_no_share
        stake_yield_proposal=eth_yield*proposed_stake_share

        #collateral needs by NOs:
        lst_operators=lst_nodes.copy()
        collateral_need_lst = [func_collateral_need(lst_operators[i],collateral_power,collateral_base) for i in range(len(lst_operators))]
        collateral_need=sum(collateral_need_lst)

        #reth yield
        reth_yield_no_improvement = eth_yield*(1-commission)
        reth_yield = reth_yield_no_improvement*(1+additional_eth_yield_performance_factor)
        reth_yield_current = eth_yield*(1-proposed_commission)
        
        dct_out['deposit_pool'].append(deposit_pool)
        dct_out['commission'].append(commission)
        dct_out['share_no_commission'].append(share_no_commission)
        dct_out['share_stake_commission'].append(1-share_no_commission)
        dct_out['stake_commission'].append((1-share_no_commission)*commission)
        dct_out['no_commission'].append(share_no_commission*commission)
        dct_out['eth_yield_no_commssion'].append(eth_yield_no_commssion)
        dct_out['eth_yield_stakers_commission'].append(eth_yield_stakers_commission)
        dct_out['rpl_yield_stakers'].append(eth_yield_stakers_commission*eth_price_init/rpl_price_init/rpl_stake_init)
        dct_out['rpl_yield_stakers_proposal'].append(stake_yield_proposal*eth_price_init/rpl_price_init/rpl_stake_init)
        dct_out['eth_yield_no_proposal'].append(eth_yield_no_proposal)
        dct_out['reth_yield'].append(reth_yield)
        dct_out['reth_yield_current'].append(reth_yield_current)
        dct_out['reth_yield_improvement_commission'].append((reth_yield_no_improvement-reth_yield_current))
        dct_out['reth_yield_improvement_performance'].append((reth_yield-reth_yield_no_improvement))
        # protocol_earnings = ((1+eth_yield)**(1/365)-1)*reth_tvl_initial
```

### RPL staking yield 

```python
#assume stable values for 1 year, loop through different deposit pool sizes and RPL stake ratios
dct_out = {
    'deposit_pool':[],
    'commission':[],
    'share_no_commission':[],
    'eth_yield_no_commssion':[],
    'eth_yield_stakers_commission':[],
    'eth_yield_no_proposal':[],
    'rpl_yield_stakers':[],
    'rpl_yield_stakers_proposal':[],
    'rpl_stake':[],
    'rpl_stake_ratio':[],
    'share_stake_commission':[],
    'stake_commission':[],
    'no_commission':[],
}

for rpl_stake in np.arange(num_validators_initial*min_stake_requirement,rpl_supply_init*0.9,1000000):
    for deposit_pool in [0,0.2*max_deposit,0.5*max_deposit,0.8*max_deposit,max_deposit]:
    
        

        #total commission and NO- and RPL staker share of commission based on deposit pool size
        protocol_earnings = eth_yield*reth_tvl_initial
        share_no_commission = func_share_no(deposit_pool, min_no_commission_share, max_no_commission_share, kink_no_share,max_deposit)
        commission = func_comission(deposit_pool, min_commission, max_commission, min_deposit, max_deposit)
        share_no_commission = func_share_no(deposit_pool, min_no_commission_share, max_no_commission_share, kink_no_share, max_deposit)
        eth_yield_no_commssion = eth_yield*share_no_commission*commission
        eth_yield_stakers_commission=eth_yield*(1-share_no_commission)*commission
        #ETH income is swapped to RPL for stakers
        rpl_income_stakers = protocol_earnings*(1-share_no_commission) *commission*eth_price_init/rpl_price_init

        #values for original proposal
        eth_yield_no_proposal=eth_yield*proposed_no_share
        stake_yield_proposal=protocol_earnings*proposed_stake_share*eth_price_init/rpl_price_init #voter share
        

        #collateral needs by NOs:
        lst_operators=lst_nodes.copy()
        collateral_need_lst = [func_collateral_need(lst_operators[i],collateral_power,collateral_base) for i in range(len(lst_operators))]
        collateral_need=sum(collateral_need_lst)
        
        
        
        dct_out['deposit_pool'].append(deposit_pool)
        dct_out['commission'].append(commission)
        dct_out['share_no_commission'].append(share_no_commission)
        dct_out['share_stake_commission'].append(1-share_no_commission)
        dct_out['stake_commission'].append((1-share_no_commission)*commission)
        dct_out['no_commission'].append(share_no_commission*commission)
        dct_out['rpl_stake'].append(rpl_stake)
        dct_out['rpl_stake_ratio'].append(rpl_stake/rpl_supply_init)
        dct_out['eth_yield_no_commssion'].append(eth_yield_no_commssion)
        dct_out['eth_yield_stakers_commission'].append(eth_yield_stakers_commission)
        dct_out['rpl_yield_stakers'].append(rpl_income_stakers/rpl_stake)
        dct_out['rpl_yield_stakers_proposal'].append(stake_yield_proposal/rpl_stake)
        dct_out['eth_yield_no_proposal'].append(eth_yield_no_proposal)
        
        
    
```

# Simulation over time

```python

dct_out = {
    'deposit_pool':[],
    'commission':[],
    'day':[],
    'eth_yield_no':[],
    'eth_yield_no_proposal':[],
    'rpl_yield_stakers':[],
    'eth_yield_stakers_proposal':[],
    'rpl_buy_pressure_cumulative':[],
    'rpl_price':[],
    'rpl_eth_price':[],
    'rpl_supply':[],
    'eth_price':[],
    'reth_tvl':[],
    'num_validators':[],
    'num_operators':[],
    'no_commission_proposal_eth':[],
    'eth_yield_no_proposal_compareable':[],
    'no_commission_eth':[],
    'no_apy_eth_based_vs_solo':[],
    'no_apy_eth_based_vs_solo_proposal':[],
    'stake_commission_proposal_eth':[],
    'commission_stake_eth':[],
    'collateral_need':[],
    'rpl_stake':[],
    'rpl_stake_min':[],
    'stake_ratio':[],
}

avg_validators_per_operator = sum(lst_nodes)/len(lst_nodes)
#pre simulate prices so there are the same for all deposit pool sizes
if do_price_change:
    lst_price_changes=[]
    for day in range(simulation_length):
        lst_price_changes.append(np.random.normal(mu_rpl/365,sigma_rpl/np.sqrt(365)))

for deposit_pool in [0,0.2*max_deposit,0.5*max_deposit,0.8*max_deposit,max_deposit]:
    
    #reset every 
    lst_operators=lst_nodes.copy()
    collateral_need_init = sum([func_collateral_need(lst_operators[i],collateral_power,collateral_base) for i in range(len(lst_operators))])
    new_operator=0
    rpl_eth_price_init = rpl_price_init/eth_price_init
    
    for day in range(simulation_length):
        if day==0:
            reth_tvl = reth_tvl_initial
            commission_stake_rpl_cs = 0
            num_validators = num_validators_initial
            num_validators_prev = num_validators_initial
            additional_rpl_stake = additional_rpl_stake_init
            rpl_eth_price = rpl_eth_price_init
            rpl_price = rpl_price_init
            eth_price=eth_price_init
        else:
            num_validators_prev = num_validators
            reth_tvl_tmp = reth_tvl*(1+growth_rate_tvl)**(1/365)
            num_validators = reth_tvl_tmp // 32 
            reth_tvl = 32*num_validators
            additional_rpl_stake=additional_rpl_stake*(1-stake_rate_deduction)**(1/365)
            if do_price_change:
                price_change=lst_price_changes[day]
            else:
                price_change = 0 
            rpl_eth_price=rpl_eth_price*np.exp(price_change)
            rpl_price = rpl_eth_price*eth_price
        
        #add new validators and over time also operators that run exactly avg_validators_per_operator validators
        new_validators_needed = num_validators - num_validators_prev
        
        if num_validators > sum(lst_operators):
            if new_validators_needed<=avg_validators_per_operator:
                new_operator += new_validators_needed
                if new_operator>=avg_validators_per_operator: #wait until avg validator count is reached
                    lst_operators.append(new_operator)
                    new_operator=0
            else:
                tmp = 0
                while tmp < new_validators_needed:
                #random.sample(lst_operators,1)[0] #not used as drawing this skews too much towards smaller operators over time
                    new_operator = min(avg_validators_per_operator, new_validators_needed-tmp)
                    lst_operators.append(new_operator)
                    tmp+=new_operator
        
        #RPL inflation should have marginal impact on stake ratio
        rpl_supply = rpl_supply_init*(1+rpl_inflation)**(day/365)

        #new collateral need, instead of sum([func_collateral_need(lst_operators[i],collateral_power,collateral_base) for i in range(len(lst_operators))]), we just scale collateral_need_init to avoid the impact in mix of validators / node operators
        collateral_need = collateral_need_init * reth_tvl/reth_tvl_initial
        # sum([func_collateral_need(lst_operators[i],collateral_power,collateral_base) for i in range(len(lst_operators))])
        collateral_borrow = reth_tvl-collateral_need
            
        share_eth_borrow = collateral_borrow/reth_tvl
        share_eth_borrow_proposal = (32-pool_size_eth)/32
        
        rpl_stake_min = min_stake_requirement*num_validators
        if rpl_stake_min*(1+additional_rpl_stake)/rpl_supply > max_stake_ratio:
            if rpl_stake_min/rpl_supply < max_stake_ratio:
                while rpl_stake_min*(1+additional_rpl_stake)/rpl_supply > max_stake_ratio:
                    additional_rpl_stake = additional_rpl_stake*(1-stake_rate_deduction)
            else:
                additional_rpl_stake = 0

            
        
        rpl_stake = rpl_stake_min*(1+additional_rpl_stake)
        rpl_stake_ratio = rpl_stake/rpl_supply

        protocol_earnings = ((1+eth_yield)**(1/365)-1)*reth_tvl
        
        commission = func_comission(deposit_pool, min_commission, max_commission, min_deposit, max_deposit)
        
        commission_eth = commission*share_eth_borrow*protocol_earnings

        share_no_commission = func_share_no(deposit_pool, min_no_commission_share, max_no_commission_share, kink_no_share,max_deposit)
        no_commission_eth  =commission_eth*share_no_commission
        

        

        #node operator income and yield
        no_income_eth = (1-share_eth_borrow)*protocol_earnings
        no_income_eth_proposal = (1-share_eth_borrow_proposal)*protocol_earnings
        no_apy_eth_based = (1+(no_commission_eth + no_income_eth)/(collateral_need))**365-1
        no_apy_eth_based_vs_solo = no_apy_eth_based/eth_yield

        #RPL staker income and yield
        commission_stake_eth = commission_eth-no_commission_eth
        #swap commission to RPL
        commission_stake_rpl = commission_stake_eth*eth_price/rpl_price

        stake_apy_rpl = (1+commission_stake_rpl/rpl_stake)**(365)-1

        
        
        #metrics from original proposal
        no_commission_proposal_eth=proposed_no_share*protocol_earnings*share_eth_borrow_proposal
        stake_commission_proposal_eth=proposed_stake_share*protocol_earnings*share_eth_borrow_proposal

        no_apy_eth_based_proposal_orig = (1+(no_commission_proposal_eth+no_income_eth_proposal)/(num_validators*pool_size_eth))**365-1
        no_apy_eth_based_proposal_collat = (1+(no_commission_proposal_eth+no_income_eth)/(collateral_need))**365-1
        no_apy_eth_based_vs_solo_proposal = no_apy_eth_based_proposal_orig/eth_yield
        
        stake_apy_eth_proposal = (1+stake_commission_proposal_eth/(rpl_stake*rpl_price/eth_price))**365-1
        


        dct_out['deposit_pool'].append(deposit_pool)
        dct_out['day'].append(day)
        dct_out['eth_yield_no'].append(no_apy_eth_based)
        dct_out['eth_yield_no_proposal'].append(no_apy_eth_based_proposal_orig)
        dct_out['eth_yield_no_proposal_compareable'].append(no_apy_eth_based_proposal_collat)
        dct_out['rpl_yield_stakers'].append(stake_apy_rpl)
        dct_out['eth_yield_stakers_proposal'].append(stake_apy_eth_proposal)
        dct_out['rpl_stake_min'].append(rpl_stake_min)
        dct_out['rpl_stake'].append(rpl_stake)
        dct_out['stake_ratio'].append(rpl_stake_ratio)
        commission_stake_rpl_cs+=commission_stake_rpl
        dct_out['rpl_buy_pressure_cumulative'].append(commission_stake_rpl_cs)
        dct_out['rpl_price'].append(rpl_price)
        dct_out['rpl_eth_price'].append(rpl_eth_price)
        dct_out['eth_price'].append(eth_price)
        dct_out['reth_tvl'].append(reth_tvl)
        dct_out['commission'].append(commission)
        dct_out['no_commission_proposal_eth'].append(no_commission_proposal_eth)
        dct_out['no_commission_eth'].append(no_commission_eth)
        dct_out['no_apy_eth_based_vs_solo'].append(no_apy_eth_based_vs_solo)
        dct_out['no_apy_eth_based_vs_solo_proposal'].append(no_apy_eth_based_vs_solo_proposal)
        dct_out['stake_commission_proposal_eth'].append(stake_commission_proposal_eth)
        dct_out['commission_stake_eth'].append(commission_stake_eth)
        dct_out['rpl_supply'].append(rpl_supply)
        dct_out['num_validators'].append(num_validators)
        dct_out['num_operators'].append(len(lst_operators))
        dct_out['collateral_need'].append(collateral_need)

        
```
